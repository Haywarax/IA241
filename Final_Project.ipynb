# Background

Covid-19 is an infectious disease caused by the SARS-CoV-2 virus.

Most people infected with the virus will experience mild to moderate respiratory illness and recover without requiring special treatment. However, some will become seriously ill and require medical attention. Older people and those with underlying medical conditions like cardiovascular disease, diabetes, chronic respiratory disease, or cancer are more likely to develop serious illness. Source: [World Health Organization](https://www.who.int/health-topics/coronavirus#tab=tab_1)

## An overview of the data for India

### This is the data collected by the [European Centre for Disease Prevention and Control](https://www.ecdc.europa.eu/en/publications-data/download-todays-data-geographic-distribution-covid-19-cases-worldwide) on India's Covid cases and population.

```python
india_data = df.loc[ df['countriesAndTerritories']=='India']
india_data[:40]
```

### A comparison between India's Population and Total number of Covid cases and Total Deaths

```python
india_data.mean()['popData2019']
```

India's Total population as of 2019.

```python
india_data.sum()['cases']
```

Total Number of Cases in India.

```python
india_data.sum()['deaths']
```

Total number of deaths in India.

## An Overview of Top 15 Countries sorted by Population

```python
sum_pop_per_country = df.groupby('countriesAndTerritories').mean()['popData2019']
sum_pop_per_country.nlargest(15).plot.bar()
```

## Top 15 Countries by Cases

```python
sum_cases_per_country = df.groupby('countriesAndTerritories').sum()['cases']
sum_cases_per_country.nlargest(15).plot.bar()
```

The two above graphs indicate that India is second in both population and number of cases however other countries positions change greatly.

## Visualization of Indian Cases and Deaths from Covid by Month

```python
india_data.plot.scatter(x='cases',y='deaths',c='month')
```

The above scatterplot indicates that with the exception of a couple outliers, there is a strong correlation between number of cases and number of deaths in India.

### Visualization of Cases by Month

```python
sum_cases_per_day = india_data.groupby('month').mean()['cases']
sum_cases_per_day.plot.bar()
```

The worst month for India in terms of cases was September. The number of deaths is strongly correlated with the number of cases as seen in the previous graph.

# Conclusion and Limitations

According to the data available to the European Centre for Disease Prevention and Control, India has handled Covid-19 relatively well. Despite having over 1.3 billion citizens, less than 10 million cases have been reported equalling less than 1% of the population. However this data may be limited by a lack of testing and reporting and may not accurately represent the situation on the ground.
